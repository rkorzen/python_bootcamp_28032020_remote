```python
# Wygenerować taki sygnał, który będzie się skłądał z puntków mierzonych co 5 minut w jakimś tak okresie czasu:

n = 24 * 12 
dti = pd.date_range("2018-01-01", periods=n, freq="5T")
values = np.sin(np.arange(n))
df = pd.DataFrame({"y": values}, index=dti )
df.plot()
```

```python
# napisz funkcje, która bedzie generowala data frame z losowymi wartościami dla zadane N

def random_df(N):
    x = np.arange(N)
    y = np.random.randint(-100, 100, N)
    df = pd.DataFrame({"y": y}, index = x)
    return df

df = random_df(100)
df.plot()
#  df 1000 wierszy. 
```

```python
def random_df(N):
    
    x = np.arange(N)
    y = np.sin(x) + np.random.uniform(0, 100, N)
    df = pd.DataFrame({"y":y}, index=x)
    return df
random_df(100).plot()
```
